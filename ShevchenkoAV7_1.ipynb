```python
import pandas as pd

def class_movies(row):
    if row['rating'] <= 2.0:
        return 'низкий рейтинг'
    elif 2.0 < row['rating'] <= 4.0:
        return 'средний рейтинг'
    elif row['rating'] == 4.5 or ['rating'] == 5.0:
        return 'высокий рейтинг'
    else:
        return 'Вне рейтинга'

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')[['movieId','rating']]
merged = ratings.merge(movies, on = 'movieId', how = 'left')


merged['class'] = merged.apply(class_movies, axis = 1)
merged.head()
```
