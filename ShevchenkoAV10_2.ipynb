```python
import pandas as pd
import numpy as np

def print_stats(col_name, low_q = 0.01, up_q = 0.99):
    print(col_name)
    print()
    print('Выбросы через межквартильный размах')
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound0 = q1 - (1.5 * iqr) 
    upper_bound0 = q3 + (1.5 * iqr)
    remove_surgery_outliers0 = df[df[col_name].between(lower_bound0, upper_bound0, inclusive='both')]
    print(df[~df[col_name].between(lower_bound0, upper_bound0, inclusive='both')][col_name])
    print('Разница со средним без выбросов ', df[col_name].mean() - remove_surgery_outliers0[col_name].mean())
    print('Разница с медианой без выбросов ', df[col_name].median() - remove_surgery_outliers0[col_name].median())
    print()
    print(f'Выбросы через {low_q}-{up_q} квантиль')
    lower_bound1 = df[col_name].quantile(low_q) 
    upper_bound1 = df[col_name].quantile(up_q)
    remove_surgery_outliers1 = df[df[col_name].between(lower_bound1, upper_bound1, inclusive='both')]
    print(df[~df[col_name].between(lower_bound1, upper_bound1, inclusive='both')][col_name])
    print('Разница среднего со средним без выбросов ', df[col_name].mean() - remove_surgery_outliers0[col_name].mean())
    print('Разница медианы с медианой без выбросов ', df[col_name].median() - remove_surgery_outliers0[col_name].median())
    print('Максимум ', df[col_name].max())
    print('Новый Максимум0 ', remove_surgery_outliers0[col_name].max())
    print('Новый Максимум1 ', remove_surgery_outliers1[col_name].max())
    print('Минимум ', df[col_name].min())
    print('Новый Минимум0 ', remove_surgery_outliers0[col_name].min())
    print('Новый Минимум1 ', remove_surgery_outliers1[col_name].min())
    print('Размах ', df[col_name].max() - df[col_name].min())
    print('Новый Размах0 ', remove_surgery_outliers0[col_name].max() - remove_surgery_outliers0[col_name].min())
    print('Новый Размах1 ', remove_surgery_outliers1[col_name].max() - remove_surgery_outliers1[col_name].min())
    print('Среднеарифметическое ', df[col_name].mean())
    print('Новое Среднеарифметическое0 ', remove_surgery_outliers0[col_name].mean())
    print('Новое Среднеарифметическое1 ', remove_surgery_outliers1[col_name].mean())
    print('Мода ', df[col_name].mode()[0])
    print('Новая Мода0 ', remove_surgery_outliers0[col_name].mode()[0])
    print('Новая Мода1 ', remove_surgery_outliers1[col_name].mode()[0])
    print('Медиана ', df[col_name].median())
    print('Новая Медиана0 ', remove_surgery_outliers0[col_name].median())
    print('Новая Медиана1 ', remove_surgery_outliers1[col_name].median())
    print('Второй квартиль (50% квантиль) ', df[col_name].quantile())
    print('Новый Второй квартиль0 (50% квантиль) ', remove_surgery_outliers0[col_name].quantile())
    print('Новый Второй квартиль1 (50% квантиль) ', remove_surgery_outliers1[col_name].quantile())
    print('Первый квартиль (25% квантиль) ', df[col_name].quantile(0.25))
    print('Новый Первый квартиль0 (25% квантиль) ', remove_surgery_outliers0[col_name].quantile(0.25))
    print('Новый Первый квартиль1 (25% квантиль) ', remove_surgery_outliers1[col_name].quantile(0.25))
    print('Третий квартиль (75% квантиль) ', df[col_name].quantile(0.75))
    print('Новый Третий квартиль0 (75% квантиль) ', remove_surgery_outliers0[col_name].quantile(0.75))
    print('Новый Третий квартиль1 (75% квантиль) ', remove_surgery_outliers1[col_name].quantile(0.75))
    print('Межквартильный размах ', df[col_name].quantile(0.75) - df[col_name].quantile(0.25))
    print('Новый Межквартильный размах0 ', remove_surgery_outliers0[col_name].quantile(0.75) - remove_surgery_outliers0[col_name].quantile(0.25))
    print('Новый Межквартильный размах1 ', remove_surgery_outliers1[col_name].quantile(0.75) - remove_surgery_outliers0[col_name].quantile(0.25))
    print('СКО (среднее отклонение) ', df[col_name].std())
    print('Новое СКО (среднее отклонение)0 ', remove_surgery_outliers0[col_name].std())
    print('Новое СКО (среднее отклонение)1 ', remove_surgery_outliers1[col_name].std())
    print('Дисперсия ', df[col_name].var())
    print('Новая Дисперсия0 ', remove_surgery_outliers0[col_name].var())
    print('Новая Дисперсия1 ', remove_surgery_outliers1[col_name].var())

url = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'
col_names = ['surgery', # Качественный (проводилась хирургическая операция или нет)
             'age', # Качественный (моложе 6 месяцев или старше)
             'hospital_number', # Качественный (идентификатор лошади)
             'rectal_temperature', # Количественный дискретный (значение градусника)
             'pulse', # Количественный дискретный
             'respiratory_rate', # Количественный дискретный
             'temperature_of_extremities', # Качественный (категории температуры не в очерёдности градусов)
             'peripheral_pulse', 'mucous_membranes', 'capillary_refill_time',
             'pain', 'peristalsis', 'abdominal_distension', 'nasogastric_tube', 'nasogastric_reflux',
             'nasogastric_reflux_ph', 'rectal_examination', 'abdomen', 'packed_cell_volume', 'total_protein',
             'abdominocentesis_appearance', 'abdomcentesis_total_protein', 'outcome', 'surgical_lesion',
             'type_of_lesion1', 'type_of_lesion2', 'type_of_lesion3', 'cp_data']
df = pd.DataFrame()
try:
    df = pd.read_csv(url, names = col_names, na_values = '?')
finally:
    if df.empty:
        df = pd.read_csv('horse_data.csv', names = col_names, na_values = '?')
df = df[['surgery', 'age', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'pain', 'outcome']]
df.head()
```
