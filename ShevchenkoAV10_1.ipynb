```python
import pandas as pd
import numpy as np

def print_stats(col_name):
    print(col_name)
    print('Максимум ', df[col_name].max())
    print('Минимум ', df[col_name].min())
    print('Размах ', df[col_name].max() - df[col_name].min())
    print('Среднеарифметическое ', df[col_name].mean())
    print('Мода ', df[col_name].mode()[0])
    print('Медиана ', df[col_name].median())
    print('Второй квартиль (50% квантиль) ', df[col_name].quantile())
    print('Первый квартиль (25% квантиль) ', df[col_name].quantile(0.25))
    print('Третий квартиль (75% квантиль) ', df[col_name].quantile(0.75))
    print('Межквартильный размах ', df[col_name].quantile(0.75) - df[col_name].quantile(0.25))
    print('СКО (среднее отклонение) ', df[col_name].std())
    print('Дисперсия ', df[col_name].var())

url = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'
col_names = ['surgery', # Качественный (проводилась хирургическая операция или нет)
             'age', # Качественный (моложе 6 месяцев или старше)
             'hospital_number', # Качественный (идентификатор лошади)
             'rectal_temperature', # Количественный дискретный (значение градусника)
             'pulse', # Количественный дискретный
             'respiratory_rate', # Количественный дискретный
             'temperature_of_extremities', # Качественный (категории температуры не в очерёдности градусов)
             'peripheral_pulse', 'mucous_membranes', 'capillary_refill_time',
             'pain', 'peristalsis', 'abdominal_distension', 'nasogastric_tube', 'nasogastric_reflux',
             'nasogastric_reflux_ph', 'rectal_examination', 'abdomen', 'packed_cell_volume', 'total_protein',
             'abdominocentesis_appearance', 'abdomcentesis_total_protein', 'outcome', 'surgical_lesion',
             'type_of_lesion1', 'type_of_lesion2', 'type_of_lesion3', 'cp_data']
df = pd.DataFrame()
try:
    df = pd.read_csv(url, names = col_names, na_values = '?')
finally:
    if df.empty:
        df = pd.read_csv('horse_data.csv', names = col_names, na_values = '?')
df = df[['surgery', 'age', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'pain', 'outcome']]
df.head()
```

```python
print_stats('surgery')
print('Вывод: Большинству лошадей проводили хирургическую операцию')
```

```python
print_stats('age')
print('Вывод: Большинство лошадей выборки являются взрослыми - больше 6 месяцев')
print('Примечание: в описании говорилось о значениях 1 и 2, в предоставленном файле - 1 и 9')
```

```python
print_stats('rectal_temperature')
print('Вывод: Большинство лошадей имеют температуру чуть выше нормы, но незначительно - 0.7 до градусов цельсия, т.к.орма 37.8, б')
```

```python
print_stats('pulse')
print('Вывод: Большинство лошадей либо имели болезненные повреждения или находились в состоянии циркулярного шока - пульс выше нормы (которая 30-40)')
print('Вывод: атлетически сложенных лошадей не попадалось')
print('Примечание: для более точных результатов скорее всего необходимо учитывать корреляцию с показателем pain')
```

```python
print_stats('respiratory_rate')
print('Вывод: подтвердилось примечание из описания - слишком большие колебания по сравнению с нормой (8-10)')
```

```python
print_stats('temperature_of_extremities')
print('Вывод: большинство лошадей имеют пониженную (cool, а не cold) температуру конечностей, что может говорить о шоке')
print('Примечание: для более точных результатов скорее всего необходимо учитывать корреляцию с показателями rectal_temperature и pain')
```

```python
print_stats('pain')
print('Вывод: оценка боли скорее всего имеет нормальное распределение')
print('Примечание: если другие столбцы имеют корреляцию с этим, можно наверное заполнить пропуски для тех столбцов')
```

```python
print_stats('outcome')
print('Вывод: Большинство лошадей выжили после лечения')
print('Примечание: для более точных результатов скорее всего необходимо учитывать корреляцию с показателем pain, точнее со значениями 4 и тем более 5')
```
