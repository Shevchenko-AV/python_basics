```python
import pandas as pd

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль' ],
    'Северо-Запад': ['петербург', 'псков', 'мурманск' ],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск' ]
}

cities = []
for values in geo_data.values():
    for city in values:
        cities.append(city)
keywords = pd.read_csv('keywords.csv')

def region_in_keyword(row):
    words = row['keyword'].split(' ')
    has_region = False
    for word in words:
        if word in cities:
            for items in geo_data.items():
                if word in items[1]:
                    has_region = True
                    return items[0]
    if not has_region:
        return 'undefined'

keywords['region'] = keywords.apply(region_in_keyword, axis = 1)
print(keywords.head())
keywords[(keywords['region'] != 'undefined')]
```
